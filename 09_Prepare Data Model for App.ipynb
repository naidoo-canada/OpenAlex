```python
server_name = 'mdrid-1.cfgquvzxu6rh.ca-central-1.rds.amazonaws.com'
database_name = 'MDRID_1'
username_ = 'admin'
pwd = '7jLkMAyKgESupoJiueft'

connection_url = sa.engine.URL.create(
    "mysql+pymysql",
    host=server_name,
    username=username_,
    password=pwd,
    database=database_name
)
engine = create_engine(connection_url)
```

Create table:

```python
table_name = 'concepts_by_pub_' + country_code + '_first_auth_since2017_APP'

metadata_obj = sa.MetaData()
table = sa.Table(
    table_name,                                        
    metadata_obj,                                    
    sa.Column('pub_id', sa.types.NVARCHAR(length=255), primary_key=True),  
    sa.Column('concept', sa.types.NVARCHAR(length=255)),                    
    sa.Column('concept_level', sa.Integer),                
)
  
metadata_obj.create_all(engine)
```
