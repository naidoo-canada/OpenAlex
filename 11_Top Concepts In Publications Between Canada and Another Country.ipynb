In response to BII and others, try to generate a list to show the top concepts for academic publications that Canadian and Spanish researchers collaborated on. 

```python
import pandas as pd
```

```python
dataDir = '/Users/rnaidoo/Documents/Canada-Secure/GAC/2022_MDRID/Projects_data/OpenAlex/'
```

#  

Load data:

```python
country_code = 'ES'
country_code_collab = 'CA'

#Set-up directories
dataDir_first_auth = dataDir + 'works_' + country_code + '_first_auth/'
dataDir_col = dataDir_first_auth + country_code_collab + '_col/'
#Inverse
dataDir_first_auth_inv = dataDir + 'works_' + country_code_collab + '_first_auth/'
dataDir_col_inv = dataDir_first_auth_inv + country_code + '_col/'
```

```python
df_pubs = pd.read_csv(dataDir_col + 'works_' + country_code + '_first_auth_' + country_code_collab + '_col_since2017.csv')
print(len(df_pubs))
df_pubs.head()
```

```python
df_pubs_inv = pd.read_csv(dataDir_col_inv + 'works_' + country_code_collab + '_first_auth_' + country_code + '_col_since2017.csv')
print(len(df_pubs_inv))
df_pubs_inv.head()
```

```python
df_conc = pd.read_csv(dataDir_first_auth + 'concepts_by_pub_' + country_code + '_first_auth_since2017.csv')
print(len(df_conc))
df_conc.head()
```

```python
df_conc_inv = pd.read_csv(dataDir_first_auth_inv + 'concepts_by_pub_' + country_code_collab + '_first_auth_since2017.csv')
print(len(df_conc_inv))
df_conc_inv.head()
```

Isolate publications that have collaborators of interest:

```python
df_pubs2 = df_pubs.loc[df_pubs['collaborators_of_interest'] != '{}']
print(len(df_pubs2))
df_pubs2.head()
```

```python
df_pubs2_inv = df_pubs_inv.loc[df_pubs_inv['collaborators_of_interest'] != '{}']
print(len(df_pubs2_inv))
df_pubs2_inv.head()
```

Isolate by concept level 0:

```python
concept_level = 0
```

```python
df_conc2 = df_conc.loc[df_conc['concept_level'] == concept_level]
print(len(df_conc2))
```

```python
df_conc2_inv = df_conc_inv.loc[df_conc_inv['concept_level'] == concept_level]
print(len(df_conc2_inv))
```

Merge concepts with publications:

```python
df_pubs3 = df_pubs2.merge(df_conc2, how='right', left_on='id', right_on='pub_id')
df_pubs3 = df_pubs3[df_pubs3['id'].notnull()]
print(len(df_pubs3))
df_pubs3.head()
```

```python
df_pubs3_inv = df_pubs2_inv.merge(df_conc2_inv, how='right', left_on='id', right_on='pub_id')
df_pubs3_inv = df_pubs3_inv[df_pubs3_inv['id'].notnull()]
print(len(df_pubs3_inv))
df_pubs3_inv.head()
```

```python
df_pubs4 = pd.concat([df_pubs3, df_pubs3_inv])
print(len(df_pubs4))
df_pubs4.head()
```

Group by concept:

```python
df_pubs5 = df_pubs4[['concept', 'cited_by_count']]
df_pubs5['publications'] = 1
print(len(df_pubs5))
df_pubs5.head()
```

```python
df_result = df_pubs5.groupby(['concept']).sum().sort_values('cited_by_count', ascending=False)
df_result
```

```python
df_result.to_csv(dataDir_col + 'works_by_concept_' + str(concept_level) + '_' + country_code + '-' + country_code_collab + '.csv')
```
