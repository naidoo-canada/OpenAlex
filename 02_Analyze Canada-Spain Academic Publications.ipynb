Process all Canadian first author publications:

```python
df1 = df_can_load.copy(deep=True)
df1['first_author'] = pd.Series()
df1['first_auth_orcid'] = pd.Series()
df1['first_auth_inst'] = pd.Series()
df1['first_auth_country'] = pd.Series()
df1['collaborators_of_interest'] = pd.Series()
df1.head()
```

```python
for i in range(0, len(df1)):
    auth_list_ = df1.iloc[i]['authorships']
    auth_list = ast.literal_eval(auth_list_)
    
    #Get first author details
    if auth_list[0]['author_position'] == 'first':
        if auth_list[0]['institutions'][0]['country_code'] == 'CA':
            first_auth = auth_list[0]['author']['display_name']
            first_auth_orcid = auth_list[0]['author']['orcid']
            first_auth_inst = auth_list[0]['institutions'][0]['display_name']
            first_auth_country = auth_list[0]['institutions'][0]['country_code']
            
    #Get collaborator authors' details
    collab_list = {}
    df_collabs = pd.DataFrame()
    for auth in auth_list:
        #auth_insts = auth['institutions']
        for inst in auth['institutions']:
            if inst['country_code'] == 'ES':
                col_auth = auth['author']['display_name']
                col_auth_orcid = auth['author']['orcid']
                col_auth_inst = inst['display_name']
                col_auth_country = inst['country_code']
                
                col_auth_dict = {
                    'col_auth_orcid': col_auth_orcid,
                    'col_auth_inst': col_auth_inst,
                    'col_auth_country': col_auth_country
                }
                
                collab_list[col_auth] = col_auth_dict
    
    #Add columns
    df1.iloc[i,26] = first_auth
    df1.iloc[i,27] = first_auth_orcid
    df1.iloc[i,28] = first_auth_inst
    df1.iloc[i,29] = first_auth_country
    df1.iloc[i,30] = str(collab_list)
```

#  

#### Determine the top institutions participating in CA-ES academic collaborations over the past 10 years, by citations on collaborative published works:

Top-Canadian Institutions:

```python
df_can_load = pd.read_csv(dataDir + 'works_can_first_auth_esp_col_proc.csv')
df_can_load['publications'] = 1
df_can_load.head()
```

```python
df_top_can_inst = df_can_load[['cited_by_count', 'publications', 'first_auth_inst']].groupby(['first_auth_inst']).sum()
df_top_can_inst = df_top_can_inst.sort_values('cited_by_count', ascending=False)
df_top_can_inst = df_top_can_inst.rename(columns={
    'cited_by_count': 'by citations',
    'publications': 'by publications'
})
df_top_can_inst.index.names = ['top Canadian institutions collaborating with Spainish researchers']
df_top_can_inst#.head(10)
```

Top-Spanish Institutions:

```python
df_esp_load = pd.read_csv(dataDir + 'works_esp_first_auth_can_col_proc.csv')
df_esp_load['publications'] = 1
df_esp_load.head()
```

```python
df_top_esp_inst = df_esp_load[['cited_by_count', 'publications', 'first_auth_inst']].groupby(['first_auth_inst']).sum()
df_top_esp_inst = df_top_esp_inst.sort_values('cited_by_count', ascending=False)
df_top_esp_inst = df_top_esp_inst.rename(columns={
    'cited_by_count': 'by citations',
    'publications': 'by publications'
})
df_top_esp_inst.index.names = ['top Spanish institutions collaborating with Canadian researchers']
df_top_esp_inst#.head(10)
```

#  

#### Determine top academics:

```python
df_top_ac = df_can_load[['cited_by_count', 'publications', 'first_author']].groupby(['first_author']).sum()
df_top_ac = df_top_ac.sort_values('cited_by_count', ascending=False)
df_top_ac2 = df_top_ac.merge(df_can_load[['first_author', 'first_auth_inst', 'first_auth_orcid']], how='left', on='first_author')
df_top_ac2 = df_top_ac2.rename(columns={
    'cited_by_count': 'by citations',
    'publications': 'by publications',
    'first_auth_inst': 'institution',
    'first_auth_orcid': 'orcid'
})
df_top_ac2 = df_top_ac2.set_index('first_author')
df_top_ac2.index.names = ['top Canadian academics collaborating with Spainish researchers']
df_top_ac2#.head(10)
```

```python
df_top_ac = df_esp_load[['cited_by_count', 'publications', 'first_author']].groupby(['first_author']).sum()
df_top_ac = df_top_ac.sort_values('cited_by_count', ascending=False)
df_top_ac2 = df_top_ac.merge(df_esp_load[['first_author', 'first_auth_inst', 'first_auth_orcid']], how='left', on='first_author')
df_top_ac2 = df_top_ac2.drop_duplicates(subset='first_author')
df_top_ac2 = df_top_ac2.rename(columns={
    'cited_by_count': 'by citations',
    'publications': 'by publications',
    'first_auth_inst': 'institution',
    'first_auth_orcid': 'orcid'
})
df_top_ac2 = df_top_ac2.set_index('first_author')
df_top_ac2.index.names = ['top Spanish academics collaborating with Canadian researchers']
df_top_ac2.head(10)
```
