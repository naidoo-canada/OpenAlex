#### Functions:

```python
def construct_query_string_200(endpoint, filters):
    
    filter_param = f'filter={",".join(filters)}'
    filtered_works_url = f'https://api.openalex.org/{endpoint}?{filter_param}&per-page=200'
    print(f'complete URL:\n{filtered_works_url}')
    
    return filtered_works_url
```

```python
def construct_query_string_cursor(endpoint, filters, cursor='*'):
    
    filter_param = f'filter={",".join(filters)}'
    filtered_works_url = f'https://api.openalex.org/{endpoint}?{filter_param}&per-page=200&cursor={cursor}'
    print(filtered_works_url)
    
    return filtered_works_url
```

```python
def isolate_can_esp_first_auth_pubs(df_resp):

    can_first_indicies = []
    esp_first_indicies = []
    for i in range(0, len(df_resp)):
        authorships_dict = df_resp['authorships'][i]
        for j in range(0, len(authorships_dict)):
            if authorships_dict[j]['author_position'] == 'first': 
                first_auth_insts = authorships_dict[j]['institutions']
                if len(first_auth_insts) > 0:
                    for k in range(0, len(first_auth_insts)): 
                        if len(first_auth_insts[k]) > 0:
                            if first_auth_insts[k]['country_code'] == 'CA':
                                can_first_indicies.append(i)
                            elif first_auth_insts[k]['country_code'] == 'ES':
                                esp_first_indicies.append(i)
                        
                                df_can_first = df_resp.iloc[can_first_indicies]
                                df_esp_first = df_resp.iloc[esp_first_indicies]

                                return df_can_first, df_esp_first
```

#  

#### Process all results using cursor pagination:

```python
endpoint = 'works'
filters = ['institutions.country_code:CA', 'institutions.country_code:ES', 'publication_year:>2012']

#Start cursor pagination
filtered_works_url = construct_query_string_cursor(endpoint=endpoint, filters=filters)
response = requests.get(filtered_works_url)
response_count = response.json()['meta']['count']
print('Total number of results: ' + str(response_count))
pages = round(np.floor(response_count/200))
remainder = np.remainder(response_count, 200)
print('Processing ' + str(pages) + ' additional pages...')
df_response = pd.DataFrame(response.json()['results'])
df_can, df_esp = isolate_can_esp_first_auth_pubs(df_response)
df_can.to_csv(dataDir + 'works_can_first_auth.csv', index=False)
df_esp.to_csv(dataDir + 'works_esp_first_auth.csv', index=False)
next_cursor = response.json()['meta']['next_cursor']

#Loop through next pages
for i in range(0, pages):
    filtered_works_url = construct_query_string_cursor(endpoint=endpoint, filters=filters, cursor=next_cursor)
    response = requests.get(filtered_works_url)
    df_response = pd.DataFrame(response.json()['results'])
    df_can_load = pd.read_csv(dataDir + 'works_can_first_auth.csv')
    df_esp_load = pd.read_csv(dataDir + 'works_esp_first_auth.csv')
    df_can, df_esp = isolate_can_esp_first_auth_pubs(df_response)
    df_can = pd.concat([df_can_load, df_can])
    df_esp = pd.concat([df_esp_load, df_esp])
    df_can.to_csv(dataDir + 'works_can_first_auth.csv', index=False)
    df_esp.to_csv(dataDir + 'works_esp_first_auth.csv', index=False)
    next_cursor = response.json()['meta']['next_cursor']
    #print(next_cursor)
    
```

#### Group-by institutions:

Determine the top institutions participating in CA-ES academic collaborations over the past 10 years, by citations on collaborative published works:

```python
endpoint = 'works'
filters = ['institutions.country_code:CA', 'institutions.country_code:ES', 'publication_year:>2012']
filtered_works_url = construct_query_string_200(endpoint=endpoint, filters=filters)
```

```python
url = filtered_works_url + '&group_by=authorships.institutions.id'
url
```

```python
response = requests.get(url)
```

```python
response.json()['meta']['count']
```

```python
df_resp = pd.DataFrame(response.json()['group_by'])
df_resp.head()
```

```python
df_inst_CA_ES = pd.concat([df_inst_CA, df_inst_ES])
df_inst_CA_ES
```

```python
df_top_CA_ES_inst_cite = df_resp.merge(df_inst_CA_ES, left_on='key', right_on='id')
df_top_CA_ES_inst_cite = df_top_CA_ES_inst_cite[['key_display_name', 'count', 'country_code']]
df_top_CA_ES_inst_cite = df_top_CA_ES_inst_cite.rename(columns={
    'key_display_name': 'institution',
    'count': 'citations on CA-ES collaborative work'
})
df_top_CA_ES_inst_cite
```
